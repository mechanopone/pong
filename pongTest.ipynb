```python
import pygame
pygame.init()

WIDTH,HEIGHT=700,500
WIN=pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption('Bong')

FPS=60

WHITE=(255,255,255)
BLACK=(0,0,0)

PADDLE_WIDTH,PADDLE_HEIGHT=60,100

class Paddle:
    COLOR=WHITE
    VEL=4
    def __init__(self,x,y,width,height):
        self.x=x
        self.y=y
        self.width=width
        self.height=height
    def draw(self,win):
        pygame.draw.rect( win , self.COLOR ,(self.x, self.y, self.width, self.height))
    def move(self,up=True):
        if up :
            self.y-=self.VEL
        else:
            self.y+=self.VEL

class Ball:
    MAX_VEL=5

    def __init__(self,x,y,radius):
            self.x=x
            self.y=y
            self.radius=radius
            self.x_vel= self.MAX_VEL
            self.y_vel= 0
    
def draw(win,paddles):
    win.fill(BLACK)
    for paddle in paddles:
        paddle.draw(win)
    for i in range(10,HEIGHT,HEIGHT//20):
        if i % 2==1:
            continue
        pygame.draw.rect(win,WHITE,(WIDTH//2-5,i,10,HEIGHT//20))
    pygame.display.update()

def handle_paddle(keys,left_pad,right_pad):
    if keys[pygame.K_w] and left_pad.y - left_pad.VEL>=0:
        left_pad.move(up=True)
    if keys[pygame.K_s] and left_pad.y + left_pad.VEL +left_pad.height<=HEIGHT:
        left_pad.move(up=False)

    if keys[pygame.K_UP] and right_pad.y - right_pad.VEL>=0:
        right_pad.move(up=True)
    if keys[pygame.K_DOWN] and right_pad.y + right_pad.VEL + right_pad.height<=HEIGHT:
        right_pad.move(up=False)

def main():
    run=True
    clock=pygame.time.Clock()

    left_pad=Paddle(10,HEIGHT//2-PADDLE_HEIGHT//2,PADDLE_WIDTH,PADDLE_HEIGHT)
    right_pad=Paddle(WIDTH-10-PADDLE_WIDTH,HEIGHT//2-PADDLE_HEIGHT//2,PADDLE_WIDTH,PADDLE_HEIGHT)
    while run:
        clock.tick(FPS)                             #regulate the framerate with tick function
        draw(WIN,[left_pad,right_pad])
        for event in pygame.event.get():
            if event.type ==pygame.QUIT:
                run=False
                break
        keys=pygame.key.get_pressed()
        handle_paddle(keys,left_pad,right_pad)
    pygame.quit()
if __name__=='__main__':
    main()
```
